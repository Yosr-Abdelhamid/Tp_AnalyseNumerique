On considère le polynôme $p(x) = (x - 1)^6$.

En développant on peut écrire ce même polynôme sous la forme $f(x) = x^6 - 6x^5 + 15x^4 - 20x^3 + 15x^2 - 6x + 1.$

```python
def p1(x):
    return (x-1)**6 
def p2(x):
    return x**6 - 6*x**5 + 15*x**4 - 20*x**3 + 15*x**2 - 6*x + 1
```

1) Avec Matlab évaluer $p(x)$ et $f(x)$ pour $x$ allant de $0.998$ à $1.002$ en choisissant des pas de $0.0001$.

* ### <I>Evaluation de p(x) avec python

```python
import numpy
for i in numpy.arange(0.998, 1.002, 0.001):
    print('p1({})={} '.format(i,p1(i)))
```

* ### <I>Evaluation de f(x) avec python

```python
import numpy
for i in numpy.arange(0.998, 1.002, 0.001):
    print('p2({})={} '.format(i,p2(i)))
```

2) Faire un graphique de $f(x)$ et $p(x)$. Commenter.

* ### <I>Représentation  graphique de p(x) 

```python
import matplotlib.pyplot as plt
import numpy as np
x= np.arange(0.998, 1.002, 0.0001)
plt.xlabel("x")
plt.ylabel("p(x)")
plt.grid(True)
plt.plot(x,p1(x))
```

* ### <I>Représentation  graphique de f(x) 

```python
import matplotlib.pyplot as plt
import numpy as np
x= np.arange(0.998, 1.002, 0.0001)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.plot(x,p2(x),color='r')
```

=> <I> On remarque qu'il ya une difference entre la représentation graphique de p(x) et f(x) et cette difference est du au erreurs de calcul. En effet plus le nombre d'operation augmente plus le taux d'erreur augmente. </I>
